```python
def string2num(s):
    s=list(s)
    if(len(s)==0):
        assert False
    if(s[0]=='â‚¹'):
        s=(s)[1:]
    arr=[]
    for i in range(len(s)):
        c=s[i]
        if(c==','):
            continue
        arr.append(c)
    s=''.join(arr)
    return float(s)
```

```python
class Timer:
    def __init__(self,period_of_one_candel):
        self.period_of_one_candel=period_of_one_candel
        self.time=None
    def reset(self):
        self.time=None
        return self
    def start(self):
        self.time=time.time()
        return self
    def istimeout(self):
        if(time.time()-self.time>=self.period_of_one_candel):
            return True
        else:
            return False
```

```python
class RawCandel:## time independent single candel
    def __init__(self):
        self.open=None
        self.close=None
        self.high=None
        self.low=None
        self.bottom=None
        self.top=None
        self.trend=None
        self.last=None
        self.isblank=True
    def push(self,data):
        self.last=data
        if(self.isblank):
            self.open=data
            self.close=data
            self.high=data
            self.low=data
            self.bottom=data
            self.top=data
            self.isblank=False
        else:
            self.close=data
            self.high=max(data,self.high)
            self.low=min(data,self.low)
            self.bottom=min(self.open,self.close)
            self.top=max(self.open,self.close)
        
        if(self.open<self.close):
            self.trend=+1
        else:
            self.trend=-1
```

```python
class CandelList:
    def __init__(self,no_of_candels_to_store,period_of_candel):
        self.arr=[]
        self.maxcount=no_of_candels_to_store
        self.period_of_one_candel=period_of_candel
        self.timer=Timer(self.period_of_one_candel)
        self.candel=RawCandel()
        self.candel_started=False
        self.total_candels_examined=0
    def push(self,price):
        if(not self.candel_started):
            self.timer.reset()
            self.timer.start()
            self.candel_started=True
            
        if(self.timer.istimeout()): 
            self.candel_started=False
            self.total_candels_examined+=1
            self.timer.reset()
            self.arr.append(self.candel)
            self.candel=RawCandel()
            self.candel.push(self.arr[-1].close)
            if(len(self.arr)>self.maxcount):
                self.arr.pop(0)
        else:# on going
            self.candel.push(price)
        
        return self

    
    
    def predictcrash(self,n,threshold):
        count=0
        summ=0
        if(self.total_candels_examined < n):
            return False,-1
        if(n > self.maxcount):
            assert False,-1
        for i in range(-1, -n-1, -1):
            summ+=self.arr[i].low
            count+=1
        summ/=count
        diff=summ-self.candel.low
        if(diff/summ>threshold):
            return (True,diff/summ)
        else:
            return (False,diff/summ)
        
        
    def predictanticrash(self,n):
        cnt=0
        sum=0
    
        for i in range(-1,-n-1,-1):
            try:
                sum+=self.arr[i].high
                cnt+=1
            except:
                break
        if(sum==0):
            return False
        sum/=cnt
        diff=self.candel.high-sum
        if(diff/sum>2/100):
            return True
        else:
            return False
```
