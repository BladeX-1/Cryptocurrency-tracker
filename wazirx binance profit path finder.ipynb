```python
def string2num(s): # to remove , '₹' etc
    s=list(s)
    if(len(s)==0):
        assert False
    arr=[]
    for i in range(len(s)):
        c=s[i]
        if(c in [',',' ','₹','/'] or ord('A')<=ord(c.upper())<=ord('Z')):
            continue
        arr.append(c)
    s=''.join(arr)
    return float(s)
```

```python
def inr_wc_bc_bu_wu_inr(wazirxInrDict,wazirxUsdtDict,binancePriceDict,coin):
    try:
        amount=100*0.999
        coinprice=wazirxInrDict[coin]
        coinquantity=amount/coinprice ##seems right
        #coin bought
        # here converted to binance with 0 fee
        priceUsdt=binancePriceDict[coin]*0.999 #in usdt
        totalUsdt=coinquantity*priceUsdt

        #usdt in wrx
        priceInr=wazirxInrDict['USDT']
        totalPriceInr=totalUsdt*priceInr*0.999
    except:
        return 0
    return totalPriceInr

def inr_wc1_bc1_bu_bc2_wc2_InrUsdt(wazirxInrDict,wazirxUsdtDict,binancePriceDict,coin,coin2):
    try:
        amount=100
        price=wazirxInrDict[coin]*1.001
        quantity1=amount/price

        # Here converted to binance with 0 fee
        priceUsdt=binancePriceDict[coin] #in usdt
        totalUsdt=quantity1*priceUsdt*0.999

        price2Usdt=binancePriceDict[coin2]
        totalQuantity2=totalUsdt*0.999/price2Usdt

        #coin in wrx

        #case2 through usdt
        priceUsdt=wazirxUsdtDict[coin2][1]
        totalUSDTpricetotal=totalQuantity2*0.999/priceUsdt
        totalInrThroughUsdt=totalUSDTpricetotal*0.999/wazirxInrDict['USDT']
        
        # Case1  inr based sale
        price=wazirxInrDict[coin2]
        totalINRprice=totalQuantity2*0.999/price
        
    except:
        return totalINRprice,totalInrThroughUsdt
    return totalINRprice,totalInrThroughUsdt

def inr_wu_bu_bc_wc_InrUsdt(wazirxInrDict,wazirxUsdtDict,binancePriceDict,coin):
    try:
        amount=100
        usdtPriceInINR=wazirxInrDict['USDT']
        usdtQuantity=amount*0.999/usdtPriceInINR

        # usdt in binance
        coinprice=binancePriceDict[coin]
        coinQuantity=usdtQuantity*0.999/coinprice

        #coin in wrx
        # Case1  inr based sale
        price=wazirxInrDict[coin]
        inrprice=coinQuantity*0.999*price

        #case2  usdt base
        priceInUsdt=wazirxUsdtDict[coin][1]
        usdtValue=coinQuantity*0.999*priceInUsdt
        usdtprice=wazirxInrDict['USDT']
        inrvalue=usdtValue*usdtprice*0.999
    except:
        return 0,0
    return inrprice,inrvalue
```
